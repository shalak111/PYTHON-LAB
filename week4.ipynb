```python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]

print("Fruits in the list:")
for fruit in fruits:
    print(fruit)
```

```python
# Add a new fruit
fruits.append("Grapes")

# Print the updated list
print("Updated list:", fruits)
```

```python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]

fruits.pop(2)

print("Updated list:", fruits)
```

```python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]
print("First fruit is:", fruits[0])
```

CREATE A LIST OF FIVE NUMBERS AND PRINT ALL ELEMENTS 

```python
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    print(num)
```

PRINT FIRST AND LAST ELEMENT OF A LIST 

```python
lst = [10, 20, 30, 40, 50]

print("First element:", lst[0])
print("Last element:", lst[-1])
```

ADD NUMBER 100 TO A LIST

```python
lst = [1, 2, 3, 4]
lst.append(100)

print(lst)
```

REMOVE ELEMENT 3 FROM THE LIST 

```python
lst = [1, 2, 3, 4, 5]
lst.remove(3)

print(lst)
```

FIND SUM OF ALL ELEMENTS

```python
lst = [5, 10, 15, 20]

total = sum(lst)
print("Sum:", total)
```

FIND THE MAXIMUM AND MINIMUM

```python
lst = [8, 3, 9, 1, 6]

print("Maximum:", max(lst))
print("Minimum:", min(lst))
```

CREATE A TUPLE OF 4 ELEMENTS AND PRINT IT 

```python
t = (10, 20, 30, 40)

print(t)
```

PRINT SECOND AND THIRD ELEMENT OF A TUPLE 

```python
t = (10, 20, 30, 40)

print("Second element:", t[1])
print("Third element:", t[2])
```

UNPACK TUPLE INTO VARIABLES AND PRINT THEM

```python
t = (5, 10, 15)

a, b, c = t

print(a)
print(b)
print(c)
```
